```python
import pandas as pd
import numpy as np

df = pd.read_csv("C:/Users/enriq/Desktop/country_vaccinations.csv")

df = pd.read_csv("C:/Users/enriq/Desktop/country_vaccinations.csv", encoding='utf8')
```

**Cambio de nombre de columnas**

Decidimos cambiar el nombre de las columnas que vamos a utilizar al idioma español para una mejor comprensión para futuras lecturas.

```python
 #https://www.iteramos.com/pregunta/7805/-renombrar-las-columnas-en-los-pandas-
df.rename(columns = {'country':'País'}, inplace=True)
df.rename(columns = {'date':'Fecha'}, inplace=True)
df.rename(columns = {'total_vaccinations':'Total_V'}, inplace=True)
df.rename(columns = {'people_vaccinated':'Personas_V'}, inplace=True)
df.rename(columns = {'people_fully_vaccinated':'PV_Totalmente'}, inplace=True)
df.rename(columns = {'daily_vaccinations':'V_Diarias'}, inplace=True)
df.rename(columns = {'vaccines':'Nombre_V'}, inplace=True)
```

```python
df.head()
```

**Eliminación de columnas**

Decidimos eliminar las columnas de iso_code, total_vaccination_per_hundred, people_vaccinated_per_hundred, people_fully_vaccinated_per_hundred, daily_vaccinations_per_million, source_name y source_website (1, 6, 8, 9, 10, 11, 13 y 14) ya que la información que proporcionaban no era de nuestro interés para este proyecto y no afectarán el análisis que haremos de la base de datos.


```python
#http://exponentis.es/como-eliminar-columnas-de-un-dataframe-de-python-pandas
df1 = df.drop(df.columns[[1, 6, 8, 9, 10, 11, 13, 14]], axis='columns')
df1
```

**Datos Nulos**

Todos las columnas que contengan datos nulos se mostraran con un "True" y los que no con un "False".

```python
df1.isnull()
```

```python
df1.isnull().any()
```

```python
df1.isnull().sum()
```

Para llenar la celdas vacias con ceros utilizaremos la siguiente instrucción:

```python
df_0= df1.fillna(0)
df_0
```

```python
df_0.dtypes
```

Cambiamos el orden de las columnas para que sea más fácil para nosotros leer la base de datos basado en nuestros objetivos. Por ejemplo, uno de ellos es ver cual vacuna ha sido más utilizada entonces movimos la columna que contiene el nombre de la vacunas que han sido aplicadas a la posición 2. (0= País, 1= Fecha, 2= Nombre_V)

```python
#Función cambio de columnas
#https://www.delftstack.com/es/howto/python-pandas/how-to-change-the-order-of-dataframe-columns/
dfn = df_0.reindex(columns = ['País', 'Fecha', 'Nombre_V', 'Personas_V', 'PV_Totalmente', 'Total_V','V_Diarias'])
dfn
```

**Guardar en archivo nuevo**

Guardamos en un nuevo archivo llamado "vacunas_por_pais" la base de datos contodas las modificaciones hechas.

```python
#https://www.analyticslane.com/2018/06/15/guardar-y-leer-archivos-csv-con-python/#:~:text=Para%20guardar%20un%20archivo%20CSV,CSV%20utilizando%20el%20m%C3%A9todo%20to_csv%20.
dfn.to_csv("C:/Users/enriq/Desktop/vacunas_por_pais.csv")
```
